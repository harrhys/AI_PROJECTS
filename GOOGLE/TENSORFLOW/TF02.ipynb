TENSOR CREATION 

Scalar with Value 10 </BR>
A scalar is a rank-0 tensor (a single value).

```python
import tensorflow as tf

# Create a scalar
scalar = tf.constant(10)

# Print the scalar
print("Scalar:")
print(scalar.numpy())

# Check its shape and rank
print("Shape:", scalar.shape)
print("Rank:", scalar.ndim)
```

Vector with Values [1, 2, 3, 4, 5] <BR>
A vector is a rank-1 tensor (a 1D array).

```python
# Create a vector
vector = tf.constant([1, 2, 3, 4, 5])

# Print the vector
print("\nVector:")
print(vector.numpy())

# Check its shape and rank
print("Shape:", vector.shape)
print("Rank:", vector.ndim)
```

3x3 Matrix with Values from 1 to 9<BR>
A matrix is a rank-2 tensor (a 2D array).

```python
# Create a 3x3 matrix
matrix = tf.constant([[1, 2, 3, 4], [4, 5, 6, 7], [7, 8, 9, 10]])

# Print the matrix
print("\nMatrix:")
print(matrix.numpy())

# Check its shape and rank
print("Shape:", matrix.shape)
print("Rank:", matrix.ndim)
```

2x2x2 Tensor with Values from 1 to 8 <Br>
A 2x2x2 tensor is a rank-3 tensor (a 3D array).

```python
# Create a 2x2x2 tensor
tensor = tf.constant([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])

# Print the tensor
print("\n2x2x2 Tensor:")
print(tensor.numpy())

# Check its shape and rank
print("Shape:", tensor.shape)
print("Rank:", tensor.ndim)
```
