```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data (y = 3x + 5 with some noise)
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 3 * X + 5 + np.random.randn(100, 1) * 0.5  # Adding noise

# Initialize parameters (randomly)
w = np.random.randn(1)  # Weight
b = np.random.randn(1)  # Bias
learning_rate = 0.1
epochs = 100  # Number of iterations

# Store loss values for visualization
loss_history = []

# Gradient Descent Algorithm
for _ in range(epochs):
    y_pred = w * X + b  # Predicted values
    loss = np.mean((y_pred - y) ** 2)  # Mean Squared Error (MSE)
    loss_history.append(loss)
    
    # Compute Gradients
    dw = (2 / len(X)) * np.sum(X * (y_pred - y))
    db = (2 / len(X)) * np.sum(y_pred - y)

    # Update weights using gradients
    w -= learning_rate * dw
    b -= learning_rate * db

# Plot loss reduction over epochs
plt.plot(range(epochs), loss_history, label="Loss")
plt.xlabel("Epochs")
plt.ylabel("Loss (MSE)")
plt.title("Gradient Descent - Loss Reduction Over Time")
plt.legend()
plt.show()

print(f"Optimized Weight: {w[0]:.4f}, Optimized Bias: {b[0]:.4f}")
```
